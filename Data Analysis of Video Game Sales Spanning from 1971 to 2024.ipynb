```python
# Removing missing values

data = data.dropna()
data.isnull().mean() * 100
```

```python
# Outliers detection, removal and visualization for each relevant numerical column

# Function to remove outliers using IQR
def remove_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    return df[(df[column] >= (Q1 - 1.5 * IQR)) & (df[column] <= (Q3 + 1.5 * IQR))]
```

```python
# Check for outliers and visualize for each relevant column

columns_to_check = ['critic_score', 'total_sales', 'na_sales', 'jp_sales', 'pal_sales', 'other_sales']

for column in columns_to_check:
    # Visualizing outliers before removal
    plt.figure(figsize=(15, 5))
    sns.boxplot(data[column])
    plt.title(f'Boxplot of {column} Before Outlier Removal')
    plt.show()
```

```python
# Remove outliers

data = remove_outliers_iqr(data, column)

# Visualize outliers after removal

for column in columns_to_check:
    plt.figure(figsize=(15, 5))
    sns.boxplot(data[column])
    plt.title(f'Boxplot of {column} After Outlier Removal')
    plt.show()
```

```python
# Standardizing 'genre' and 'console'

data['genre'] = data['genre'].str.lower().str.strip()
data['console'] = data['console'].str.lower().str.strip()
```

```python
# changing the release_date & last_update in datetime format

data['release_date'] = pd.to_datetime(data['release_date'], errors='coerce')
data['last_update'] = pd.to_datetime(data['last_update'], errors='coerce')
```

```python
data.dtypes
```

```python
# Descriptive statistics

data.describe()
```

```python
# Changing to numeric

data['critic_score'] = pd.to_numeric(data['critic_score'], errors='coerce')
data['total_sales'] = pd.to_numeric(data['total_sales'], errors='coerce')
data['na_sales'] = pd.to_numeric(data['na_sales'], errors='coerce')
data['jp_sales'] = pd.to_numeric(data['jp_sales'], errors='coerce')
data['pal_sales'] = pd.to_numeric(data['pal_sales'], errors='coerce')
data['other_sales'] = pd.to_numeric(data['other_sales'], errors='coerce')
```

```python
data.dtypes
```

```python
# Selecting only numeric columns for correlation matrix

numeric_data = data.select_dtypes(include=[np.number])
```

```python
# Calculating the correlation matrix

correlation_matrix = numeric_data.corr()
correlation_matrix
```

```python
# Plotting the heatmap

plt.figure(figsize=(15, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```

# Data Visualization

```python
# Creating impactful visualizations

# Total Sales by Genre

genre_sales = data.groupby('genre')['total_sales'].sum().sort_values()
plt.figure(figsize=(12, 8))
genre_sales.plot(kind='barh')
plt.title('Total Sales by Genre')
plt.xlabel('Sales (in millions)')
plt.ylabel('Genre')
plt.show()
```

```python
# Sales Trends Over Time

sales_trends = data.groupby(data['release_date'].dt.year)['total_sales'].sum()
plt.figure(figsize=(12, 6))
sales_trends.plot(kind='line')
plt.title('Total Sales Over Time')
plt.legend()
plt.xlabel('Year')
plt.ylabel('Total Sales (in millions)')
plt.show()
```

```python
# Critic Score vs. Total Sales

plt.figure(figsize=(12, 6))
sns.scatterplot(data=data, x='critic_score', y='total_sales')
plt.title('Critic Score vs. Total Sales')
plt.xlabel('Critic Score')
plt.ylabel('Total Sales (in millions)')
plt.show()
```

```python
# Regional Sales Distribution

regional_sales = data[['na_sales', 'jp_sales', 'pal_sales', 'other_sales']].sum()
regional_sales.plot(kind='pie', autopct='%1.1f%%', startangle=90)
plt.title('Regional Sales Distribution')
plt.legend()
plt.ylabel('')
plt.show()
```

```python
# Top 10 Selling Titles

top_titles = data.nlargest(10, 'total_sales')
plt.figure(figsize=(12, 6))
sns.barplot(x='total_sales', y='title', data=top_titles)
plt.title('Top 10 Selling Titles')
plt.xlabel('Sales (in millions)')
plt.ylabel('Title')
plt.show()
```

```python
# Genre Specialization by Console

console_genre = data.groupby(['console', 'genre']).size().unstack()
plt.figure(figsize=(15, 8))
sns.heatmap(console_genre, annot=True, cmap='YlGnBu')
plt.title('Console Genre Specialization')
plt.ylabel('Console')
plt.xlabel('Genre')
plt.show()
```

```python
# Boxplot of Sales by Genre

plt.figure(figsize=(15, 8))
sns.boxplot(x='genre', y='total_sales', data=data)
plt.title('Boxplot of Total Sales by Genre')
plt.xticks(rotation=45)
plt.show()
```

```python
# Distribution of Critic Scores

plt.figure(figsize=(12, 6))
sns.histplot(data['critic_score'], bins=20, kde=True)
plt.title('Distribution of Critic Scores')
plt.xlabel('Critic Score')
plt.ylabel('Frequency')
plt.show()
```

# Regional Popularity and Flops

```python
# Titles with significant disparities

data['max_region_sales'] = data[['na_sales', 'jp_sales', 'pal_sales', 'other_sales']].max(axis=1)
data['min_region_sales'] = data[['na_sales', 'jp_sales', 'pal_sales', 'other_sales']].min(axis=1)
data['disparity'] = data['max_region_sales'] - data['min_region_sales']
flop_titles = data[data['disparity'] > 1]  # Flops defined as a disparity greater than 1 million
print("Titles with significant sales disparities:\n", flop_titles[['title', 'na_sales', 'jp_sales', 'pal_sales', 'other_sales', 'disparity']])
pd.set_option('display.max_columns', None)
```
